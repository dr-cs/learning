```python
import matplotlib.pyplot as plt
import numpy
import numpy as np
import pandas

%matplotlib inline
```

```python
# Set up data

# One-letter codes for days of week in English
days = ['M', 'T', 'W', 'R', 'F', 'S', 'U']

# 7-element array of identical values (12)
milk_consumption = np.full(7, 12)
oj_consumption = np.full(7, 8)

# On the first and fifth day of the week, we use an additional 40 oz of OJ
oj_consumption[0] += 40
oj_consumption[4] += 40
```

```python
# Plot daily consumption
fig1, ax1 = plt.subplots(1, 1)

# Milk in blue with dashed line
ax1.plot(np.arange(7), milk_consumption, '--', color='blue', label='Milk')
# OJ in orange with dotted line
ax1.plot(np.arange(7), oj_consumption, ':', color='orange', label='OJ')
ax1.legend()
ax1.set_title('Milk and OJ Consumption Per Day')
ax1.set(xlim=[0, 6], xticklabels=days, xlabel='Day of Week', ylabel='Ounces')
```

```python
# Plot cumulative consumption
fig2, ax2 = plt.subplots(1, 1)

ax2.plot(np.arange(7), np.cumsum(milk_consumption), '--', color='blue', label='Milk')
ax2.plot(np.arange(7), np.cumsum(oj_consumption), ':', color='orange', label='OJ')

ax2.legend()
ax2.set_title('Cumulative Milk and OJ Consumption')
ax2.set(xlim=[0, 6], xticklabels=days, xlabel='Day of Week', ylabel='Ounces')
```
